```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("https://github.com/CSSEGISandData/COVID-19/raw/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv")
df.drop(['Province/State','Lat','Long'], axis=1, inplace=True)

us_df = df[df['Country/Region'] == "US"]
us_df = us_df.groupby('Country/Region', as_index=False).sum()
us_df = us_df.loc[:, (us_df != 0).any(axis=0)]

br_df = df[df['Country/Region'] == "Brazil"]
br_df = br_df.loc[:, (br_df != 0).any(axis=0)]

it_df = df[df['Country/Region'] == "Italy"]
it_df = it_df.loc[:, (it_df != 0).any(axis=0)]

jp_df = df[df['Country/Region'] == "Japan"]
jp_df = jp_df.loc[:, (jp_df != 0).any(axis=0)]

sk_df = df[df['Country/Region'] == "Korea, South"]
sk_df = sk_df.loc[:, (sk_df != 0).any(axis=0)]

ch_df = df[df['Country/Region'] == "China"]
ch_df = ch_df.groupby('Country/Region', as_index=False).sum()
ch_df = ch_df.loc[:, (ch_df != 0).any(axis=0)]

fr_df = df[df['Country/Region'] == "France"]
fr_df = fr_df.groupby('Country/Region', as_index=False).sum()
fr_df = fr_df.loc[:, (fr_df != 0).any(axis=0)]

ir_df = df[df['Country/Region'] == "France"]
ir_df = ir_df.groupby('Country/Region', as_index=False).sum()
ir_df = ir_df.loc[:, (ir_df != 0).any(axis=0)]
```

```
br_df.columns = range(br_df.shape[1])
us_df.columns = range(us_df.shape[1])
it_df.columns = range(it_df.shape[1])
jp_df.columns = range(jp_df.shape[1])
sk_df.columns = range(sk_df.shape[1])
ch_df.columns = range(ch_df.shape[1])
fr_df.columns = range(fr_df.shape[1])
ir_df.columns = range(ir_df.shape[1])
```

```
fig, axs = plt.subplots(1, 1, figsize=(18, 10))

axs.set_title(f"Total de casos")
axs.set_ylabel(f"Infectados")
axs.set_xlabel(f"Número de dias após primero caso confirmado")
axs.plot(br_df.iloc[0,1:],c="green",label='Brasil')
axs.plot(us_df.iloc[0,1:],c="red",label='USA')
axs.plot(it_df.iloc[0,1:],c="blue",label='Itália')
axs.plot(sk_df.iloc[0,1:],c="cyan",label='Coreia do Sul')
axs.plot(ch_df.iloc[0,1:],c="yellow",label='China')
axs.plot(fr_df.iloc[0,1:],c="brown",label='France')
axs.legend(loc='upper left',fontsize='xx-large')
```
